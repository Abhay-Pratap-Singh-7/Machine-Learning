```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[:, [3,4]].values
```

```python
import scipy.cluster.hierarchy as hc
dendogram = hc.dendrogram(hc.linkage(X, method= 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
agc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='ward')
y_hc = agc.fit_predict(X)
```

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=100, c='red', label='Cluster1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=100, c='blue', label='Cluster2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=100, c='green', label='Cluster3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s=100, c='yellow', label='Cluster4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], s=100, c='orange', label='Cluster5')
plt.show()
```

```python
agc2 = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hc2 = agc2.fit_predict(X)
```

```python
plt.scatter(X[y_hc2 == 0, 0], X[y_hc2 == 0, 1], s=100, c='red', label='Cluster1')
plt.scatter(X[y_hc2 == 1, 0], X[y_hc2 == 1, 1], s=100, c='blue', label='Cluster2')
plt.scatter(X[y_hc2 == 2, 0], X[y_hc2 == 2, 1], s=100, c='green', label='Cluster3')
plt.show()
```
